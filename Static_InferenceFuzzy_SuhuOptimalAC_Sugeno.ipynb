```python
# define a function to convert normal characters to "super" characters
def get_super(x):
    normal = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+-=()"
    super_s = "ᴬᴮᶜᴰᴱᶠᴳᴴᴵᴶᴷᴸᴹᴺᴼᴾQᴿˢᵀᵁⱽᵂˣʸᶻᵃᵇᶜᵈᵉᶠᵍʰᶦʲᵏˡᵐⁿᵒᵖ۹ʳˢᵗᵘᵛʷˣʸᶻ⁰¹²³⁴⁵⁶⁷⁸⁹⁺⁻⁼⁽⁾"
    res = x.maketrans(''.join(normal), ''.join(super_s))
    return x.translate(res)
derajat = get_super('o') # call the get_super function and assign the result to the variable 'derajat'

# define a function to generate a plot of the membership function for the Suhu Luar Ruangan
def Grafik_Suhu_Luar_Ruangan(xdingin,ydingin,xsejuk,ysejuk,xnormal,ynormal,xhangat,yhangat,xpanas,ypanas):
    plt.title("Membership Function")
    plt.xlabel("Suhu Luar Ruangan")
    plt.ylabel("Derajat Keanggotaan")
    plt.plot(xdingin,ydingin,label='Dingin')
    plt.plot(xsejuk,ysejuk,label='Sejuk')
    plt.plot(xnormal,ynormal,label='Normal')
    plt.plot(xhangat,yhangat,label='Hangat')
    plt.plot(xpanas,ypanas,label='Panas')
    plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
    plt.show()

# define a function to generate a plot of the membership function for the Suhu Dalam Ruangan
def Grafik_Suhu_Dalam_Ruangan(xsejuk,ysejuk,xnormal,ynormal,xhangat,yhangat):
    plt.title("Membership Function")
    plt.xlabel("Suhu Dalam Ruangan")
    plt.ylabel("Derajat Keanggotaan")
    plt.plot(xsejuk,ysejuk,label='Sejuk')
    plt.plot(xnormal,ynormal,label='Normal')
    plt.plot(xhangat,yhangat,label='Hangat')
    plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
    plt.show()
    
# define a function to generate a plot of the membership function for the Banyak Orang Dalam Ruangan
def Grafik_Banyak_Orang_Dalam_Ruangan(xsedikit,ysedikit,xsedang,ysedang,xbanyak,ybanyak):
    plt.title("Membership Function")
    plt.xlabel("Banyak Orang Dalam Ruangan")
    plt.ylabel("Derajat Keanggotaan")
    plt.plot(xsedikit,ysedikit,label='Sedikit')
    plt.plot(xsedang,ysedang,label='Sedang')
    plt.plot(xbanyak,ybanyak,label='Banyak')
    plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
    plt.show()
    
# define a function to generate a plot of the membership function for the Suhu Optimal AC
def Grafik_Suhu_Optimal_AC(xdingin,ydingin,xcukupd,ycukupd,xsejuk,ysejuk,xcukups,ycukups,xnormal,ynormal):
    plt.title("Membership Function")
    plt.xlabel("Suhu Optimal AC")
    plt.ylabel("Derajat Keanggotaan")
    plt.plot(xdingin,ydingin,label='Dingin')
    plt.plot(xcukupd,ycukupd,label='Cukup Dingin')
    plt.plot(xsejuk,ysejuk,label='Sejuk')
    plt.plot(xcukups,ycukups,label='Cukup Sejuk')
    plt.plot(xnormal,ynormal,label='Normal')
    plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
    plt.show()
    
# define a unique word in advance
def unique(list1):
    unique_list = []
    for x in list1:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list
```

## FUZZIFICATION
#### Input Set Determination
Based on the example above,<br />
~Suhu Luar Ruangan: 28 ,lingustic terms include: ['sejuk', 'normal']<br />
~Suhu Dalam Ruangan: 33 ,lingustic terms include: ['normal', 'hangat']<br />
~Banyak Orang Dalam Ruangan: 32 ,lingustic terms include: ['sedikit', 'sedang']<br />

```python
# Define empty lists for the linguistic terms for each input variable
him_input_slr = []
him_input_sdr = []
him_input_bodr = []

# Determine which linguistic term applies to the Suhu Luar Ruangan
if 0<=slr<=25:
    him_input_slr.append("dingin")
if 20<=slr<=30:
    him_input_slr.append("sejuk")
if 25<=slr<=35:
    him_input_slr.append("normal")
if 30<=slr<=40:
    him_input_slr.append("hangat")
if 35<=slr<=50:
    him_input_slr.append("panas")          

# Determine which linguistic term applies to the Suhu Dalam Ruangan
if 0<=sdr<=30:
    him_input_sdr.append("sejuk")
if 20<=sdr<=40:
    him_input_sdr.append("normal")
if 30<=sdr<=50:
    him_input_sdr.append("hangat")

# Determine which linguistic term applies to the Banyak Orang Dalam Ruangan
if 0<=bodr<=35:
    him_input_bodr.append("sedikit")
if 20<=bodr<=50:
    him_input_bodr.append("sedang")
if 35<=bodr<=60:
    him_input_bodr.append("banyak")

# Print the lists of linguistic terms for each input variable
print(him_input_slr)
print(him_input_sdr)
print(him_input_bodr)
```

#### Defines the Set of Input Membership Degree Values Using Formula of Triangular and Trapezodial Degree Values
Based on the example above,<br/>
~Membership Degree Values for Suhu Luar Ruangan: 28 inside ['sejuk', 'normal'] is [0.4, 0.6]<br />
~Membership Degree Values for Suhu Dalam Ruangan: 33 inside  ['normal', 'hangat'] is [0.7, 0.3]<br />
~Membership Degree Values for Banyak Orang Dalam Ruangan: 32 inside  ['sedikit', 'sedang'] is [0.2, 0.8]<br />

```python
# Define an empty list for the degree of membership values for each input values
him_y_slr = []
him_y_sdr = []
him_y_bodr = []  

# Check if the linguistic term from above is in the list of linguistic terms for each input variabels
# Suhu Luar Ruangan
if any("dingin" in s for s in him_input_slr):
    y1Dingin = max(min(1,(x1Dingin[2]-slr)/(x1Dingin[2]-x1Dingin[1])),0)
    him_y_slr.append(y1Dingin)
if any("sejuk" in s for s in him_input_slr):
    y1Sejuk = max(min((slr-x1Sejuk[0])/(x1Sejuk[1]-x1Sejuk[0]),(x1Sejuk[2]-slr)/(x1Sejuk[2]-x1Sejuk[1])),0)
    him_y_slr.append(y1Sejuk)
if any("normal" in s for s in him_input_slr):
    y1Normal = max(min((slr-x1Normal[0])/(x1Normal[1]-x1Normal[0]),(x1Normal[2]-slr)/(x1Normal[2]-x1Normal[1])),0)
    him_y_slr.append(y1Normal)
if any("hangat" in s for s in him_input_slr):
    y1Hangat = max(min((slr-x1Hangat[0])/(x1Hangat[1]-x1Hangat[0]),(x1Hangat[2]-slr)/(x1Hangat[2]-x1Hangat[1])),0)
    him_y_slr.append(y1Hangat)
if any("panas" in s for s in him_input_slr):
    y1Panas = max(min((slr-x1Panas[0])/(x1Panas[1]-x1Panas[0]),1),0)
    him_y_slr.append(y1Panas)

# Suhu Dalam Ruangan
if any("sejuk" in s for s in him_input_sdr):
    y2Sejuk = max(min(1,(x2Sejuk[2]-sdr)/(x2Sejuk[2]-x2Sejuk[1])),0)
    him_y_sdr.append(y2Sejuk)
if any("normal" in s for s in him_input_sdr):
    y2Normal = max(min((sdr-x2Normal[0])/(x2Normal[1]-x2Normal[0]),(x2Normal[2]-sdr)/(x2Normal[2]-x2Normal[1])),0)
    him_y_sdr.append(y2Normal)
if any("hangat" in s for s in him_input_sdr):
    y2Hangat = max(min((sdr-x2Hangat[0])/(x2Hangat[1]-x2Hangat[0]),1),0)
    him_y_sdr.append(y2Hangat)

# Banyak Orang Dalam Ruangan
if any("sedikit" in s for s in him_input_bodr):
    y3Sedikit = max(min(1,(x3Sedikit[2]-bodr)/(x3Sedikit[2]-x3Sedikit[1])),0)
    him_y_bodr.append(y3Sedikit)
if any("sedang" in s for s in him_input_bodr):
    y3Sedang = max(min((bodr-x3Sedang[0])/(x3Sedang[1]-x3Sedang[0]),(x3Sedang[2]-bodr)/(x3Sedang[2]-x3Sedang[1])),0)
    him_y_bodr.append(y3Sedang)
if any("banyak" in s for s in him_input_bodr):
    y3Banyak = max(min((bodr-x3Banyak[0])/(x3Banyak[1]-x3Banyak[0]),1),0)
    him_y_bodr.append(y3Banyak)

# Print the lists of Input Membership Degree Values for each input variable
print(him_y_slr)
print(him_y_sdr)
print(him_y_bodr)
```
